```python
m = folium.Map(location=[22.5205,88.3457], zoom_start=12, tiles="Stamen Terrain")

tooltip = "Click me!"

folium.Marker(
    [22.5205,88.3457], popup="<i>My Address</i>", tooltip=tooltip
).add_to(m)
m
```

```python
m.save("map.html") ##to save as html file
```

```python
m = folium.Map(location=[22.5205,88.3457], zoom_start=10, tiles="Stamen Terrain")

folium.Marker(
    location=[22.5205,88.5],
    popup="First Location",
    icon=folium.Icon(icon="cloud"),
).add_to(m)

folium.Marker(
    location=[22.5205,88.2],
    popup="Second Location",
    icon=folium.Icon(color="green"),
).add_to(m)

folium.Marker(
    location=[22.5205,88.1],
    popup="Third Location",
    icon=folium.Icon(color="red", icon="info-sign"),
).add_to(m)


m
```

```python
m = folium.Map(location=[22.5205,88.3457], zoom_start=10, tiles="Stamen Terrain")

folium.Circle(
    radius=20,
    location=[22.5623,88.3621],
    popup="GD Hospital",                ##Get an open ring in the location 
    color="crimson",
    fill=False,
).add_to(m)

folium.CircleMarker(
    location=[22.5728,88.3445],
    radius=15,
    popup="Millenium Park",
    color="#3186cc",                 ##Get a closed ring in the location 
    fill=True,
    fill_color="#3186cc",
).add_to(m)

m
```

```python
m = folium.Map(location=[22.5205,88.3457], zoom_start=10, tiles="Stamen Terrain")

m.add_child(folium.LatLngPopup())
##click any location to get the map coordinates
m
```

```python
m = folium.Map(location=[22.5205,88.3457], zoom_start=10, tiles="Stamen Terrain")

folium.Marker([22.5623,88.3621], popup="GD Hospital").add_to(m)

m.add_child(folium.ClickForMarker(popup="Waypoint")) ##you can mark at any place on the map to plot the way map


m
```
